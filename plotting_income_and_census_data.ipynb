```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
census_path = 'resources/final_census_data.csv'
income_path = 'resources/final_income_data.csv'
```

```python
census_df = pd.read_csv(census_path)
income_df = pd.read_csv(income_path)
```

```python
census_df.head()
```

```python
census_state_df = census_df[census_df['sum_level'] == 'State']
```

```python
census_state_df.head()
```

```python
income_df.head()
```

```python
sns.set(rc={'figure.figsize':(18,15)})
```

```python
sns.lineplot(data=census_state_df, x='date', y='population', hue='state')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('State Populations Over Time')
plt.savefig('census_data.png')
```

```python
sns.lineplot(data=income_df[::-1], x='date', y='income', hue='state')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Average Income Per State Over Time')
plt.savefig('income_data_2.png')
```
